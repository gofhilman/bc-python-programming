# Lists

## List Basics

List = kumpulan values yang bisa diakses hanya dengan menggunakan satu nama variable  
Values dalam list disebut *item* atau bisa juga *element*

Contoh:  
- List kosong: 
    ```python
    my_list = []
    ```
- List yang berisi 5 items: 
    ```python
    my_list = [7, 2, 2, 5, 2]
    ```

### Mengakses items dalam list



````python
my_list = [7, 2, 2, 5, 2]

print(my_list[0])
print(my_list[1])
print(my_list[3])

print("The sum of the first two items:", my_list[0] + my_list[1])
print(my_list)
````

Items dalam list bisa diganti, contoh:

````python
my_list = [7, 2, 2, 5, 2]
print(my_list)
my_list[1] = 3
print(my_list)
````

Fungsi `len` memberikan informasi banyaknya item dalam list:

````python
my_list = [7, 2, 2, 5, 2]
print(len(my_list))
````

### Menambahkan item ke dalam list

Metode `append` menambahkan item ke bagian akhir dari list, contoh:

````python
numbers = []
numbers.append(5)
numbers.append(10)
numbers.append(3)
print(numbers)
````

### Menambahkan item ke lokasi tertentu

Metode `insert` menambahkan sebuah item pada indeks yang ditentukan. Semua item yang sudah ada dalam list dengan indeks yang sama atau lebih tinggi dari indeks tersebut akan bergeser satu posisi ke kanan:



````python
numbers = [1, 2, 3, 4, 5, 6]
numbers.insert(0, 10)
print(numbers)
numbers.insert(2, 20)
print(numbers)
````

### Menghapus item dari list

Ada dua cara:
- Metode `pop` menghapus item dengan menggunakan index
- Metode `remove` menghapus item dengan menggunakan nilai konten

Contoh metode `pop`:

````python
my_list = [1, 2, 3, 4, 5, 6]

my_list.pop(2)
print(my_list)
my_list.pop(3)
print(my_list)
````

Metode `pop` juga me-return item yang dihapus:

````python
my_list = [4, 2, 7, 2, 5]

number = my_list.pop(2)
print(number)
print(my_list)
````

Contoh metode `remove`:

````python
my_list = [1, 2, 1, 2]

my_list.remove(1)
print(my_list)
my_list.remove(1)
print(my_list)
````

Jika item yang ditentukan tidak ada dalam list, maka metode `remove` akan menghasilkan error.

### Mengurutkan list

Ada dua cara:
- Metode `sort` mengurutkan dari yang terkecil sampai terbesar dengan memodifikasi list secara langsung
- Metode `sorted` membuat salinan baru urutan dari yang terkecil sampai terbesar. Dengan demikian, urutan awal tidak berubah.

````python
my_list = [2,5,1,2,4]
my_list.sort()
print(my_list)
````

````python
original = [2, 5, 1, 2, 4]
in_order = sorted(original)
print(original)
print(in_order)
````

### Maximum, minimum, dan penjumlahan

- Fungsi `max` mengembalikan item terbesar list
- Fungsi `min` mengembalikan item terkecil list
- Fungsi `sum` mengembalikan jumlah total semua item dalam list

````python
my_list = [5, 2, 3, 1, 4]

greatest = max(my_list)
smallest = min(my_list)
list_sum = sum(my_list)

print("Smallest:", smallest)
print("Greatest:", greatest)
print("Sum:", list_sum)
````

## Iterasi Definit

Kita dapat menggunakan `while` loop untuk menelusuri items dalam list, contoh:

````python
my_list = [3, 2, 4, 5, 2]

index = 0
while index < len(my_list):
    print(my_list[index])
    index += 1
````

Namun ada cara yang lebih mudah, yaitu dengan menggunakan:

### For loop

Syntax `for` loop:
```python
for <variable> in <collection>:
    <block>
```



````python
my_list = [3, 2, 4, 5, 2]

for item in my_list:
    print(item)
````

### Fungsi `range`

- Fungsi `range(n)` menyediakan rentang integer dari 0 sampai `n - 1`, contoh:

````python
for i in range(5):
    print(i)
````

- Fungsi `range(a, b)` menyediakan rentang integer dari `a` sampai `b - 1`, contoh:

````python
for i in range(3, 7):
    print(i)
````

- Fungsi `range(a, b, c)` menyediakan rentang dari `a` sampai `b - 1` dengan perubahan bernilai `c` di setiap iterasi, contoh:

````python
for i in range(1, 9, 2):
    print(i)
````

````python
for i in range(6, 2, -1):
    print(i)
````

### Membuat list dari range

Fungsi `range` mengembalikan object `range`, contoh:

````python
numbers = range(2, 7)
print(numbers)
````

Fungsi `list` akan mengubah object `range` menjadi list, contoh:

````python
numbers = list(range(2, 7))
print(numbers)
````
