# Polymorphism

Polymorphism berarti “banyak bentuk”. Dalam pemrograman, ini mengacu pada kemampuan objek berbeda untuk merespons method yang sama dengan cara yang berbeda.

```python
class Cat:
    def speak(self):
        return "Meow!"

class Dog:
    def speak(self):
        return "Woof!"

def make_sound(animal):
    print(animal.speak())

# Polymorphism in action
c = Cat()
d = Dog()

make_sound(c)  # Output: Meow!
make_sound(d)  # Output: Woof!
```

- Fungsi `make_sound()` tidak peduli apakah objeknya `Cat` atau `Dog`.
- Selama objek punya method `speak()`, fungsi akan bekerja.
- Ini disebut *polymorphism*: satu fungsi, banyak bentuk objek.

## Mengapa Polymorphism Penting?

- Membuat kode lebih fleksibel dan modular.
- Memudahkan penggunaan fungsi umum untuk berbagai tipe objek.
- Cocok untuk desain program yang scalable dan mudah dirawat.
