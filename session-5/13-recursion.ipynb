# Recursion

Sebuah function juga bisa memanggil dirinya sendiri, tetapi kita sebagai programmer harus berhati-hati saat melakukannya. Sangat mudah terjebak dalam loop tak berujung dari pemanggilan function, mirip seperti kita terjebak dalam infinite loop saat menggunakan while loop jika kita lupa menyertakan kondisi break yang tepat. Contohnya, kode berikut akan menghasilkan eror:

```python
def hello(name : str):
    print("Hello", name)
    hello(name) # function calls itself
```

## Apa arti recursion?

Dalam konteks pemrograman, recursion adalah sebuah function yang memanggil dirinya sendiri. Agar proses ini berjalan tanpa menyebabkan infinite loop, nilai argumen yang dikirim ke function tersebut harus berubah setiap kali dipanggil, sehingga rangkaian pemanggilan akan berhenti pada suatu saat nanti. Prinsip dasarnya sama seperti pada while loop: harus selalu ada kondisi berhenti, dan kondisi tersebut harus benar-benar terpenuhi selama eksekusi berlangsung. Contoh:

```python
def fill_list(numbers: list):
    """ If the length of the list is less than 10, add items to the list """
    if len(numbers) < 10:
        numbers.append(0)
        # call the function again
        fill_list(numbers)


if __name__ == "__main__":
    test_list = [1,2,3,4]
    fill_list(test_list)
    print(test_list)
```

Sebenarnya, fungsi seperti ini juga bisa dibuat dengan `while` loop biasa:

```python
def fill_list(numbers: list):
    """ If the length of the list is less than 10, add items to the list """
    while len(numbers) < 10:
        numbers.append(0)

if __name__ == "__main__":
    test_list = [1,2,3,4]
    fill_list(test_list)
    print(test_list)
```

## Recursion dan return values



```python
def factorial(n: int):
    """ The function calculates the factorial n! for n >= 0 """
    if n < 2:
        # The factorial for 0 and 1 is 1
        return 1

    # Call the function again with an argument that is one smaller
    return n * factorial(n - 1)

if __name__ == "__main__":
    # Tesing our function
    for i in range(1, 7):
        print(f"The factorial of {i} is {factorial(i)}")
```

```python
def fibonacci(n: int):
    """ The function returns the nth number in the Fibonacci sequence (1, 1, 2, 3, 5, 8 etc.); n > 0"""

    if n <= 2:
        # the first two are ones
        return 1

    # All other numbers equal the sum of the two preceding numbers in the sequence
    return fibonacci(n - 1) + fibonacci(n - 2)

# Test that everything works
if __name__ == "__main__":
    for i in range(1, 11):
        print(f"The {i}. number in the Fibonacci sequence is {fibonacci(i)}")
```
