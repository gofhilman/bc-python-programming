# Conditional Statements

## Conditional Statetement Basics

Kita bisa membuat program yang hanya dijalankan dalam situasi tertentu, contoh:

```python
age = int(input("How old are you? "))

if age > 17:
    print("You are of age!")
    print("Here's a copy of GTA6 for you.")

print("Next customer, please!")
```

### Comparison operators (Operator perbandingan)



```python
number = int(input("Please type in a number: "))

if number < 0:
    print("The number is negative.")

if number > 0:
    print("The number is positive.")

if number == 0:
    print("The number is zero.")
```

### Indentation

Suatu kode berada dalam satu blok jika indentasinya sama, contoh:

```python
password = input("Please type in a password: ")

if password == "kittycat":
    print("You knew the password!")
    print("You must be either the intended user...")
    print("...or quite an accomplished hacker.")

print("The program has finished its execution. Thanks and bye!")
```

### Boolean values dan boolean expression

Boolean values adalah nilai yang hanya berupa `True` atau `False`.  
Kode yang menghasilkan boolean values disebut boolean expression.  
Contoh:

```python
a = 3
condition = a < 5
print(condition)
if condition:
    print("a is less than 5")
```

## Percabangan dalam Conditional Statements

Kita bisa menggunakan `else` untuk membuat code block yang akan dijalankan ketika kondisi dalam `if` bernilai salah. Contoh:

```python
number = int(input("Please type in a number: "))

if number < 0:
    print("The number is negative")
else:
    print("The number is positive or zero")
```

Jika ada lebih dari dua pilihan, kita bisa menggunakan percabangan `elif` yang merupakan singkatan dari "else if".



Contoh:

```python
goals_home = int(input("Home goals scored: "))
goals_away = int(input("Away goals scored: "))

if goals_home > goals_away:
    print("The home team won!")
elif goals_away > goals_home:
    print("The away team won!")
else:
    print("It's a tie!")
```

```python
print("Holiday calendar")
date = input("What is the date today? ")

if date == "Dec 26":
    print("It's Boxing Day")
elif date == "Dec 31":
    print("It's Hogmanay")
elif date == "Jan 1":
    print("It's New Year's Day")

print("Thanks and bye.")
```

## Menggabungkan Conditions

### Logical operators / boolean operators





```python
number = int(input("Please type in a number: "))
if not (number >= 5 and number <= 8):
    print("The number is not within the range of 5 to 8")
```

### Nested Conditionals

Conditional statements juga bisa disisipkan ke dalam conditional statements lainnya. Contoh:

```python
number = int(input("Please type in a number: "))

if number > 0:
    if number % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is negative or zero")
```
