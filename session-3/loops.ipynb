# Loops

## Loop Sederhana

```python
while True:
    code = input("Please type in your PIN: ")
    if code == "1234":
        break
    print("Incorrect...try again")

print("Correct PIN entered!")
```

### Helper variables

Contoh berikut menggunakan helper variable `attempts` untuk mencatat berapa kali pengguna telah memasukkan PIN dan helper variable `success` yang bisa diatur ke nilai `True` atau `False` berdasarkan apakah pengguna berhasil masuk atau tidak:

```python
attempts = 0

while True:
    code = input("Please type in your PIN: ")
    attempts += 1

    if code == "1234":
        success = True
        break

    if attempts == 3:
        success = False
        break

    # this is printed if the code was incorrect AND there have been less than three attempts
    print("Incorrect...try again")

if success:
    print("Correct PIN entered!")
else:
    print("Too many attempts...")
```

## Loops dengan Conditions

```python
number = int(input("Please type in a number: "))

while number < 10:
    print(number)
    number += 1

print("Execution finished.")
```

### Initialization, condition, dan update

- Initialization = penetapan nilai awal dari variabel-variabel yang digunakan dalam loop condition
- Condition = situasi yang menentukan seberapa lama loop akan dijalankan
- Update = pembaruan variabel-variabel agar setiap iterasi membawa loop semakin dekat ke titik akhirnya



```python
# Ask the user for a number 
number = int(input("Please type in a number: ")) 

# Repeat while the number is less than 10 
while number < 10: 
    
    # Print out and increment 
    print (number) 
    number += 1 
    
print ("Execution finished. ")
```

## More Loops

### Break command

`break` command digunakan untuk segera menghentikan loop.  
Dua contoh berikut mempunyai fungsi yang sama:

```python
# 1st version using the break command

sum = 0

while True:
    number = int(input("Please type in a number, -1 to exit: "))
    if number == -1:
        break
    sum += number

print (f"The sum is {sum}")
```

```python
# 2nd version without the break command

sum = 0
number = 0

while number != -1:
    number = int(input("Please type in a number, -1 to exit: "))
    if number != -1:
        sum += number

print (f"The sum is {sum}")
```

### Continue command

`continue` command menyebabkan eksekusi loop langsung melompat ke awal loop, contoh:

```python
sum = 0

while True:
    number = int(input("Please type in a number, -1 to exit: "))
    if number == -1:
        break
    if number >= 10:
        continue
    sum += number

print (f"The sum is {sum}")
```

### Nested Loops

Sama seperti conditional statements, loop juga bisa disisipkan ke dalam loop lainnya. Contoh:

```python
while True:
    number = int(input("Please type in a number: "))
    if number == -1:
        break
    while number > 0:
        print(number)
        number -= 1
```
